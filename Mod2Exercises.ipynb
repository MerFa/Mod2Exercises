### Exercise 2

```python
resDfAppDrug = dfAppDrug_sorted[dfAppDrug_sorted.first_approval >= 2011]
lsMolID = list(resDfAppDrug.molecule_chembl_id.values)

if resDfAppDrug.molecule_chembl_id.values in np.unique(resDfAppDrug.molecule_chembl_id.values): # check whether we have unique molecules
    print('We have ' + str(len(lsMolID)) + ' unique molecules.\n')

mol2targ = dict()
for molID in lsMolID:
    mol2targ[molID] = set()

for ii in range(0,len(lsMolID)):
    activities = new_client.activity.filter(molecule_chembl_id__in=lsMolID[ii]).only(['molecule_chembl_id', 'target_chembl_id'])
    print(str(ii+1) + ' of ' + str(len(lsMolID)) + ': Filtered ' + str(len(activities)) + ' targets ', end =" ")
    tmpLs = list(activities.all())
    tmpLs = [tmpLs[jj]['target_chembl_id'] for jj in range(0,len(activities))]
    tmpSet = set(tmpLs) # to get unique elements
    for targID in tmpSet:
        mol2targ[lsMolID[ii]].add(targID)
    print('... and added ' + str(len(mol2targ[lsMolID[ii]])) + ' unique targets.')
```

```python
count = 0
for keyMol,valTargs in mol2targ.items():
    count += 1
    print(keyMol, end=': ')
    for targ in valTargs:
        print(targ, end=' ')
    if valTargs==set():
        print('No targets', end='')
    print()
    if count==10:
        break
```

```python
mol2prot = mol2targ.copy()
count = 0
for key, val in mol2prot.items():
    lval = list(val)
    uniprots = set()
    count += 1
    print(str(count) + ' of ' + str(len(mol2prot)) + ': ' + str(len(lval)) + ' targets ', end="")
    chunk_size = 50
    for ii in range(0,len(val),chunk_size):
        targets = new_client.target.filter(target_chembl_id__in=lval[ii:ii+chunk_size]).only(['target_components'])
        uniprots |= {comp['accession'] for tt in targets for comp in tt['target_components']}
    print('... and ' + str(len(uniprots)) + ' uniprots.')
    mol2prot[key] = uniprots
```

```python
count = 0
for keyMol,valProts in mol2prot.items():
    count += 1
    print(keyMol, end=': ')
    for prot in valProts:
        print(prot, end=' ')
    if valProts==set():
        print('No uniprots', end='')
    print()
    if count==10:
        break
```

```python
nbOfProt = [len(mol2prot[x]) for x in mol2prot if isinstance(mol2prot[x], set)]
print(str(nbOfProt) + '\n')
# print('average: ' + str(sum(nbOfProt)/len(nbOfProt)))

import statistics
tmpMean = statistics.mean(nbOfProt)
tmpMed = statistics.median(nbOfProt)
tmpSdev = statistics.stdev(nbOfProt)
print('mean: ' + str(tmpMean))
print('median: ' + str(tmpMed))
print('standard deviation: ' + str(tmpSdev))
```

### Exercise 3

```python
import requests, sys
import json
```

```python
accProtNb = set()
for val in mol2prot.values():
    for assNb in list(val):
        if assNb != set():
            accProtNb.add(assNb)
        
print(len(accProtNb))
```

```python
strLs = list()
for assNb in accProtNb:
    requestURL = 'https://www.ebi.ac.uk/proteins/api/proteins?offset=0&size=100&accession=' + assNb
    r = requests.get(requestURL, headers={ "Accept" : "application/json"})
    if not r.ok:
      r.raise_for_status()
      sys.exit()
    responseBody = r.text
    #print(responseBody)
    aa = responseBody.find('keywords')
    if aa>0:
        bb1 = responseBody[aa:].find(',"geneLocations":')
        bb2 = responseBody[aa:].find(',"references":')
        if 0<bb1 and bb1<bb2:
            bb = bb1
        else:
            bb = bb2
        resTmp = json.loads(responseBody[aa+10:aa+bb])
        print('.',end='')
        for entries in resTmp:
            strLs.append(list(entries.values())[0])
```

```python
wordfreq = []
for w in strLs:
    wordfreq.append(strLs.count(w))
print('Word frequencies computed.')

tmpInd1 = [ii for ii, xx in enumerate(wordfreq) if xx == max(wordfreq)]
print('Most common word found.')
wordfreq_sorted = wordfreq.copy()
wordfreq_sorted.sort()
tmpInd2 = [ii for ii, xx in enumerate(wordfreq) if xx == wordfreq_sorted[-len(tmpInd1)-1]]
print('Second most common words found.\n')

kWordSet1 = set()
for ii in tmpInd1:
    kWordSet1.add(strLs[ii])
kWordSet2 = set()
for ii in tmpInd2:
    kWordSet2.add(strLs[ii])

print('The keyword\n')
for kWord in kWordSet1:
    print(kWord)
print('\nappears in ' + str(max(wordfreq)) + ' of the ' + str(len(accProtNb)) + ' proteins.\n\n')

print('The keyword\n')
for kWord in kWordSet2:
    print(kWord)
print('\nappears ' + str(wordfreq_sorted[-len(tmpInd1)-1]) + ' of the ' + str(len(accProtNb)) + ' proteins.\n\n')
```
